## Data Analysis and Visualization A* Algorithm 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats

set_matplotlib_formats("retina")
plt.rc("font", family="Malgun Gothic")
plt.rc("axes", unicode_minus = False)
plt.style.use("fivethirtyeight")

colors = ['skyblue']
```

```python
mapSizes = [20, 50, 100, 500]
obstacleRatios = [4, 10, 40, 100]
fileName = []

#make csv file names.
for mapSize in mapSizes:
    for obstacleRatio in obstacleRatios:
        fileName.append(str(obstacleRatio) + "_" + str(mapSize) + ".csv")
        
fileName
```

```python
# Check load correctly.
pd.read_csv('../csv/' + fileName[0], encoding = "utf-8")
```

# Data Preprocessing & Visualization of 4_20.csv

```python
df = pd.read_csv('../csv/' + fileName[0], encoding = "utf-8")
df["20x20"] = df["20x20"].replace("20x20",400)
df.columns = ["맵 크기", "장애물 개수", "소요시간(ms)"]
time = df["소요시간(ms)"]
time_series = time.describe().round(decimals=2)
df
```

```python
df["소요시간(ms)"].value_counts().sort_index().plot(title = "단위 (ms)", xlim= (0, 7), figsize=(8, 5), rot=0)
```

```python
g = time_series[1:].plot.bar(title = "단위 : 밀리초 (ms)", color=colors, figsize=(9,5), rot = 0)

for i in range(len(time_series) - 1):
    case_count = time_series.iloc[i+1]
    g.text(x=i-0.1, y=case_count, s=case_count)
```
